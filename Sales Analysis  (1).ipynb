### Key Performing Indicators(KPI):-

##### Core Financial KPIs

```python
total_sales       = df["Sales"].sum()
total_profit      = df["Profit"].sum()
total_orders      = df["Order ID"].nunique()
total_customers   = df["Customer ID"].nunique()
total_quantity    = df["Quantity"].sum()
profit_margin     = (total_profit / total_sales) * 100
avg_order_value   = total_sales / total_orders
avg_profit_order  = total_profit / total_orders
return_on_sales   = (total_profit / total_sales) * 100

print("\n FINANCIAL KPIs")
print(f"  Total Revenue          : ${total_sales:>12,.2f}")
print(f"  Total Profit           : ${total_profit:>12,.2f}")
print(f"  Profit Margin          : {profit_margin:>11.2f}%")
print(f"  Return on Sales (ROS)  : {return_on_sales:>11.2f}%")
print(f"  Avg Order Value (AOV)  : ${avg_order_value:>12,.2f}")
print(f"  Avg Profit per Order   : ${avg_profit_order:>12,.2f}")
print(f"  Total Units Sold       : {total_quantity:>12,}")
```

#### Customer KPIs 

```python
revenue_per_customer = total_sales / total_customers
orders_per_customer  = total_orders / total_customers
top_segment          = df.groupby("Segment")["Sales"].sum().idxmax()
top_segment_pct      = df.groupby("Segment")["Sales"].sum().max() / total_sales * 100

print("\n CUSTOMER KPIs")
print(f"  Total Customers              : {total_customers:>10,}")
print(f"  Total Orders                 : {total_orders:>10,}")
print(f"  Revenue per Customer (ARPU)  : ${revenue_per_customer:>10,.2f}")
print(f"  Orders per Customer          : {orders_per_customer:>10.2f}")
print(f"  Top Segment                  : {top_segment:>10} ({top_segment_pct:.1f}% of sales)")
```

```python
# ── Product KPIs ─────────────────────────────────────────────
top_category    = df.groupby("Category")["Sales"].sum().idxmax()
top_subcategory = df.groupby("Sub-Category")["Sales"].sum().idxmax()
top_product     = df.groupby("Product Name")["Sales"].sum().idxmax()
loss_sku_count  = df[df["Profit"] < 0]["Product Name"].nunique()
total_skus      = df["Product Name"].nunique()

print("\n📦 PRODUCT KPIs")
print(f"  Total Unique Products    : {total_skus:>10,}")
print(f"  Loss-making Products     : {loss_sku_count:>10,} ({loss_sku_count/total_skus*100:.1f}%)")
print(f"  Top Category             : {top_category:>10}")
print(f"  Top Sub-Category         : {top_subcategory:>10}")
print(f"  Best Selling Product     : {top_product[:40]}")
```

#### Operational KPIs

```python
avg_ship_days    = df["Ship Days"].mean()
on_time_pct      = (df["Ship Days"] <= 5).sum() / len(df) * 100   # within 5 days
discount_rate    = (df["Discount"] > 0).sum() / len(df) * 100
avg_discount_val = df[df["Discount"] > 0]["Discount"].mean() * 100
high_disc_loss   = df[(df["Discount"] >= 0.4) & (df["Profit"] < 0)].shape[0]

print("\n🚚 OPERATIONAL KPIs")
print(f"  Avg Shipping Days        : {avg_ship_days:>10.2f} days")
print(f"  On-Time Delivery (≤5d)   : {on_time_pct:>10.1f}%")
print(f"  % Orders with Discount   : {discount_rate:>10.1f}%")
print(f"  Avg Discount (when used) : {avg_discount_val:>10.1f}%")
print(f"  High-Disc Loss Orders    : {high_disc_loss:>10,}")
```

#### YoY Growth KPIs

```python
yearly_sales   = df.groupby("Year")["Sales"].sum()
yearly_profit  = df.groupby("Year")["Profit"].sum()
yearly_orders  = df.groupby("Year")["Order ID"].nunique()

yoy_sales      = yearly_sales.pct_change() * 100
yoy_profit     = yearly_profit.pct_change() * 100
yoy_orders     = yearly_orders.pct_change() * 100

print("\n📈 YEAR-OVER-YEAR GROWTH KPIs")
print(f"  {'Year':<8} {'Sales':>12} {'YoY%':>8} {'Profit':>12} {'YoY%':>8} {'Orders':>8} {'YoY%':>8}")
print("  " + "-" * 68)
for yr in yearly_sales.index:
    print(f"  {yr:<8} ${yearly_sales[yr]:>11,.0f} {yoy_sales[yr]:>7.1f}%"
          f" ${yearly_profit[yr]:>11,.0f} {yoy_profit[yr]:>7.1f}%"
          f" {yearly_orders[yr]:>8,} {yoy_orders[yr]:>7.1f}%")
```

#### Regional KPIs

```python
region_kpi = df.groupby("Region").agg(
    Sales    = ("Sales",    "sum"),
    Profit   = ("Profit",   "sum"),
    Orders   = ("Order ID", "nunique"),
    Customers= ("Customer ID", "nunique")
).assign(
    Margin   = lambda x: (x["Profit"] / x["Sales"] * 100).round(2),
    AOV      = lambda x: (x["Sales"] / x["Orders"]).round(2)
).sort_values("Sales", ascending=False)

print("\n🗺️  REGIONAL KPIs")
print(region_kpi.to_string())
```

#### Category KPIs

```python
cat_kpi = df.groupby("Category").agg(
    Sales    = ("Sales",    "sum"),
    Profit   = ("Profit",   "sum"),
    Quantity = ("Quantity", "sum"),
    Orders   = ("Order ID", "nunique")
).assign(
    Margin      = lambda x: (x["Profit"] / x["Sales"] * 100).round(2),
    Sales_Share = lambda x: (x["Sales"] / x["Sales"].sum() * 100).round(2)
).sort_values("Sales", ascending=False)

print("\n🏷️  CATEGORY KPIs")
print(cat_kpi.to_string())
```

#### Segment KPIs 

```python
seg_kpi = df.groupby("Segment").agg(
    Sales     = ("Sales",       "sum"),
    Profit    = ("Profit",      "sum"),
    Customers = ("Customer ID", "nunique"),
    Orders    = ("Order ID",    "nunique")
).assign(
    Margin       = lambda x: (x["Profit"] / x["Sales"] * 100).round(2),
    Revenue_Share= lambda x: (x["Sales"] / x["Sales"].sum() * 100).round(2),
    CLV          = lambda x: (x["Sales"] / x["Customers"]).round(2)   # Customer Lifetime Value proxy
).sort_values("Sales", ascending=False)

print("\n👤 SEGMENT KPIs (incl. CLV proxy)")
print(seg_kpi.to_string())
```

```python
# -------OUTLIER DETECTION----------
from scipy import stats
for col in ["Sales", "Profit"]:
    Q1, Q3 = df[col].quantile(0.25), df[col].quantile(0.75)
    IQR    = Q3 - Q1
    iqr_outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]

    z_scores = np.abs(stats.zscore(df[col]))
    z_outliers = df[z_scores > 3]

    print(f"\n{col}:")
    print(f"  IQR method  → {len(iqr_outliers)} outliers ({len(iqr_outliers)/len(df)*100:.2f}%)")
    print(f"  Z-score (>3)→ {len(z_outliers)} outliers ({len(z_outliers)/len(df)*100:.2f}%)")
    print(f"  Range: ${df[col].min():.2f}  to  ${df[col].max():.2f}")
```

#### HYPOTHESIS TESTING

```python
# ------ Chi-Square Test------------
# Chi-Square Test: Check relationship between Segment and Ship Mode


# Step 1: Create a contingency table
contingency = pd.crosstab(df["Segment"], df["Ship Mode"])

# Step 2: Apply Chi-Square test
chi2, p_val, dof, expected = chi2_contingency(contingency)

# Step 3: Print hypothesis
print("H0: Segment and Ship Mode are independent")

# Step 4: Print test results
print("Chi-Square Value:", round(chi2,4))
print("Degrees of Freedom:", dof)
print("P-value:", p_val)

# Step 5: Interpretation
if p_val < 0.05:
    print("Result: Reject H0 → Segment and Ship Mode are related")
else:
    print("Result: Fail to Reject H0 → Segment and Ship Mode are independent")

# Step 6: Show observed frequencies
print("\nObserved Frequencies:")
print(contingency)
```
